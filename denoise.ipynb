```python
import glob,cv2;
import warnings
warnings.filterwarnings('ignore')
import numpy as np;
from PIL import Image
import matplotlib.pyplot as plt
orig_image=sorted(glob.glob('/media/DiskE/DnCNN-keras-master/Untitled Folder/50/DnCNN-keras-master/snapshot/test50/original/*.*',recursive=True))
noisy_img=sorted(glob.glob('/media/DiskE/DnCNN-keras-master/Untitled Folder/50/DnCNN-keras-master/snapshot/test50/noisy/*.*',recursive=True))
dncnn_img=sorted(glob.glob('/media/DiskE/DnCNN-keras-master/Untitled Folder/50/DnCNN-keras-master/snapshot/test50/dncnn/*.*',recursive=True))
rdcnn_img=sorted(glob.glob('/media/DiskE/DnCNN-keras-master/Untitled Folder/50/DnCNN-keras-master/snapshot/test50/rdcnn/*.*',recursive=True))
```

```python
w=np.random.randint(0,230);
h=np.random.randint(0,350);
#w=33
#h=140
i=np.random.randint(0,1);
print(i)
print(w,h)

img1=noisy_img[i]
img2=orig_image[i]
img3=dncnn_img[i]
img4=rdcnn_img[i]


j=64

img1=cv2.imread(img1,0)
img2=cv2.imread(img2,0)
img3=cv2.imread(img3,0)
img4=cv2.imread(img4,0)

img5=img1[w:w+j,h:h+j]
img6=img2[w:w+j,h:h+j]
img7=img3[w:w+j,h:h+j]
img8=img4[w:w+j,h:h+j]
#print(type(img5))
#print(img1.shape)
height, width= img1.shape

cv2.rectangle(img1, (h, w), ( h+j,w+j), (255,255,0), 1)
cv2.rectangle(img2, (h, w), ( h+j,w+j), (0,255,0), 1)
cv2.rectangle(img3, (h, w), ( h+j,w+j), (0,255,0), 1)
cv2.rectangle(img4, (h, w), ( h+j,w+j), (0,255,0), 1)

patch=128
height=height-patch
width=width-patch
#print(img5.shape,img1.shape)

img5 = cv2.resize(img5 , (j*2,j*2))
img6 = cv2.resize(img6 , (j*2,j*2))
img7 = cv2.resize(img7 , (j*2,j*2))
img8 = cv2.resize(img8 , (j*2,j*2))
img1[height:height+patch,width:width+patch]=img5
img2[height:height+patch,width:width+patch]=img6
img3[height:height+patch,width:width+patch]=img7
img4[height:height+patch,width:width+patch]=img8
#cv2.rectangle(img, (h, w), ( h+j,w+j), (255,0,0), 2)
fig = plt.figure(figsize=(20,20))
ax=plt.subplot(1,4,1)
#plt.axis('off')
ax.imshow(img1,cmap='gray')
ax.set_title("Original")
ax.set_xlabel('PSNR').set_color('red')
plt.xticks([])
plt.yticks([])

ax=plt.subplot(1,4,2)
#plt.axis('off')
ax.imshow(img2,cmap='gray')
ax.set_title("Original")
ax.set_xlabel('PSNR').set_color('red')
plt.xticks([])
plt.yticks([])

ax=plt.subplot(1,4,3)
#plt.axis('off')
ax.imshow(img3,cmap='gray')
ax.set_title("Original")
ax.set_xlabel('PSNR').set_color('red')
plt.xticks([])
plt.yticks([])

ax=plt.subplot(1,4,4)
#plt.axis('off')
ax.imshow(img4,cmap='gray')
ax.set_title("Original")
ax.set_xlabel('PSNR').set_color('red')
plt.xticks([])
plt.yticks([])
plt.subplots_adjust(wspace=0)
plt.show()
```
